```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"


city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)


merged_df = pd.merge(ride_df, city_df , how = 'left' )
merged_df
```

```python
avg_fare = merged_df.groupby('city', as_index=False)['fare'].mean()
avg_fare.head()
```

```python
num_of_rides =  merged_df.groupby('city')['city'].count()
num_of_rides.head()
```

```python
num_of_drivers = merged_df.groupby('city' , as_index=False)['driver_count'].sum()
num_of_drivers.head()
```

```python
fare_sum = merged_df.groupby('type' , as_index=False)['fare'].sum()
fare_sum['percentage'] = (fare_sum['fare']/fare_sum['fare'].sum())*100
print(fare_sum)

colors = ['Gold', 'LightSkyBlue' , 'LightCoral']
labels=fare_sum['type'] 
plt.pie(
    fare_sum['fare'],
    labels=fare_sum['type'],
    shadow=False,
    colors=colors,
    explode=(0, 0, 0.1),
    startangle=0,
    autopct='%1.1f%%')
plt.title("percentage of Total Fares by City Type", bbox={'facecolor':'0.9', 'pad':8})
plt.legend(labels,loc=4)
plt.show()
```

```python
rides = merged_df.groupby('type' ,as_index=False)['city'].count()
print(rides)

labels=fare_sum['type']
plt.pie(
    rides['city'],
    labels= rides['type'],
    shadow=True,
    colors=colors,
    explode=(0, 0, 0.1),
    startangle=90,
    autopct='%1.1f%%')
plt.title("percentage of Total Rides by City Type", bbox={'facecolor':'0.9', 'pad':8})
plt.legend(labels,loc=3)
plt.show()
```

```python
drivers = merged_df.groupby('type' ,as_index=False )['driver_count'].sum()
print(drivers)

labels=fare_sum['type']
plt.pie(
    drivers['driver_count'],
    labels= rides['type'],
    shadow=True,
    colors=colors,
    explode=(0, 0, 0.1),
    startangle=120,
    autopct='%1.1f%%')
plt.title("percentage of Total Drivers by City Type", bbox={'facecolor':'0.9', 'pad':8})
plt.legend(labels,loc=1)
plt.show()
```
